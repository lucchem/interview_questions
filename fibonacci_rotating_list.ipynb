```python
for N in range(500):
    i = 2
    nums = [0,1,1]
    while nums[i % 3] < N:
        nums[(i+1) % 3] = nums[i % 3] + nums[(i-1) % 3]
        i += 1
    print(N,nums[i % 3])
```
